```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
x.shape,y.shape
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y1, "r-",linewidth=0.5)
plt.plot(x, y2, "bo", markersize=3)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis("equal")
plt.title("SIN AND COS")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1, "r-",linewidth=0.5)
ax.plot(x, y2, "bo", markersize=3)
ax.set_xlim(0, 10)
ax.set_ylim(-4, 4)
ax.axis("equal")
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1, "r-",linewidth=0.5)
ax.plot(x, y2, "bo", markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 4), title="Cos dan Sin", xlabel="X", ylabel="Y")
ax.axis('equal');
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1, "r-",linewidth=0.5, label="Sin")
ax.plot(x, y2, "bo", markersize=3, label="Cos")
ax.set(xlim=(0, 10), ylim=(-2, 2), title="Cos dan Sin", xlabel="X", ylabel="Y")
ax.axis('equal');
ax.legend(loc="lower center")
```

# Scatter Plot

```python
df = pd.read_csv("../03 - Matplotlib Intro/data/circle_data.csv", sep=";")
df
```

```python
x = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(x, y)
plt.axis("equal")
```

```python
plt.scatter(x, y, c="r", s=(10))
plt.axis("equal")
```

```python
plt.scatter(x, y, c=rad, s=rad * 90, cmap="seismic")
plt.axis("equal")
```

# Histogram

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=25);
```

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=25, density=1);
```

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=25, density=1, cumulative=True);
```

# SubPlot

### Matlab Style

```python
plt.subplot(231)
plt.hist(numbers, bins=80, density=1)

plt.subplot(233)
plt.hist(numbers, bins=80, density=1)

plt.subplot(235)
plt.scatter(x, y, c='r', s=10)
plt.axis("equal")

plt.subplot(236)
plt.hist(numbers, bins=80, density=1);
```

### Matplotlib style

```python
fg, ax = plt.subplots(2, 3, figsize=(15, 10))
ax[0, 0].hist(numbers, bins=80, density=1);
ax[0, 1].axis("off")
ax[0, 2].hist(numbers, bins=80, density=1);

ax[1, 0].scatter(x, y, c='r', s=10)

ax[0, 0].hist(numbers, bins=80, density=1);

ax[0, 0].hist(numbers, bins=80, density=1);
```
